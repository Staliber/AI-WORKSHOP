```python
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns={'Selling Area Size (sq ft)':'Sales Area(sq ft)'},inplace=True)
```

```python
df['Sales Area(sq ft)']
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
```

```python
df['Order Date']
```

```python
df['Month']=df['Order Date'].dt.month
```

```python
df['Month']
```

```python
df['Day']=df['Order Date'].dt.day
```

```python
df['Day']
```

```python
monthly_sales=df.groupby('Month')['Sales Profit'].sum()
```

```python
monthly_sales
```

```python
monthly_sales.plot(kind='line')
```

```python
monthly_sales.plot(kind='bar')
```

```python
df.groupby('City')['Sales Profit'].sum()
```

```python
orders_per_year=df.groupby(df['Order Date'].dt.year)['Order ID'].nunique()
```

```python
orders_per_year
```

```python
df["Product Sale"] = df["Price"] * df["Quantity"]
```

```python
avg_sales_product=df.groupby('Product ID')['Product Sale'].mean()
```
