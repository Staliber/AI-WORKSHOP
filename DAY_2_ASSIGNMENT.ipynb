12.Generate 20 random integers and compute mean,meadian and standard deviation

```python
rand=np.random.rand(20)
print(rand)
print(np.mean(rand))
print(np.median(rand))
print(np.std(rand))
```

13.Using boolean masking,print values greater than 50

```python
arr3=np.array([12,34,45,67,79,94])
mask=arr3 > 50 #boolean masking is used to avoid writing explicit loops
print(arr3[mask])
```

14.Replace values less than 50 with 50

```python
arr4=np.array([5,12,27,34,45,67,79,94,101,123])
mask=arr4 < 50
arr4[mask]=50
print(arr4)
```
